```python
import numpy as np
import tensorflow as tf
import random

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
train_dataset = np.array([[0,0],[0,1],[1,0],[1,1]]).astype(np.float32)
train_labels = np.array([[0],[1],[1],[0]]).astype(np.float32)
```

```python
num_hidden_nodes1 = 128
num_hidden_nodes2 = 128


def tf_prediction(X,W1,b1,W2,b2,W3,b3):
    h1 = tf.nn.relu(tf.matmul(X,W1) + b1)
    h2 = tf.nn.relu(tf.matmul(h1,W2) + b2)
    return tf.matmul(h2,W3) + b3

graph = tf.Graph()
with graph.as_default():
    X = tf.placeholder(tf.float32, shape=[4,2])
    y = tf.placeholder(tf.float32, shape=[4,1])

    # W1 = tf.Variable(tf.truncated_normal([2,num_hidden_nodes1], 0, 0.1))
    W1 = tf.Variable(tf.random_uniform([2,num_hidden_nodes1], -0.1, 0.1))
    b1 = tf.Variable(tf.zeros([num_hidden_nodes1]))

    # W2 = tf.Variable(tf.truncated_normal([num_hidden_nodes1,num_hidden_nodes2], 0, 0.1))
    W2 = tf.Variable(tf.random_uniform([num_hidden_nodes1,num_hidden_nodes2], -0.1, 0.1))    
    b2 = tf.Variable(tf.zeros([num_hidden_nodes2]))
    
    # W3 = tf.Variable(tf.truncated_normal([num_hidden_nodes2,1], 0, 0.1))
    W3 = tf.Variable(tf.random_uniform([num_hidden_nodes2,1], -0.1, 0.1))    
    b3 = tf.Variable(tf.zeros([1]))

    y_predict = tf_prediction(X,W1,b1,W2,b2,W3,b3)
    
    # Minimize the mean squared errors
    loss = tf.reduce_mean(tf.square(y_predict - y))
    optimizer = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
```

```python
with tf.Session(graph=graph) as sess:
    sess.run(tf.initialize_all_variables())
    
    init_weights1, init_biases1, init_weights2, init_biases2, init_weights3, init_biases3 = sess.run([W1, b1, W2, b2, W3, b3]) 

    for step in range(401):        
        feed_dict = {X : train_dataset, y : train_labels}

        _, l, predictions = sess.run([optimizer, loss, y_predict], feed_dict=feed_dict)

        if step % 40 == 0:
            print('Loss at step %d: %f ' % (step,l))

print("Final predictions:")
print(predictions)
```
